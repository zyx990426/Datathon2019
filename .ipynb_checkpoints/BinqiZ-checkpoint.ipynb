```python
collisions_cleaned = collisions[['street_name', 'street_type']]
collisions_cleaned.head()
```

```python
counts_cleaned = counts[['street_name', 'total_count']]
counts_cleaned.head()
```

```python
full = collisions['street_type'].unique()
full
```

```python
codes = set()
for i in counts_cleaned["street_name"]:
    temp = i.split()
#     for n in temp:
    if len(temp[-1]) == 2:
        codes.add(temp[-1])
    elif len(temp[-1]) == 4 and temp[-1] in full:
        codes.add(temp[-1])
```

```python
brief_to_full = {'AV': 'AVENUE', 'BL': 'BOULEVARD', 'CR': 'CIRCLE', 'CT': 'COURT', 'DR': 'DRIVE', 'HY': 'HIGHWAY', 'LN': 'LANE',
                 'ML': 'MALL', 'PL': 'PLACE', 'PY': 'PARKWAY', 'PT': 'POINT', 'RD': 'ROAD', 'ST': 'STREET', 'TL': 'TRAIL', 'TR': 'TERRACE', 'WY': 'WAY', 'null': np.nan}
```

```python
street_type = []
for row in counts_cleaned.iterrows():
    if len(row[1][0].split()[-1]) == 2 and row[1][0].split()[-1] in brief_to_full.keys():
        temp = row[1][0].split()[-1]
        street_type.append(temp)
    else:
        street_type.append('null')
street_type
```

```python
counts_cleaned['street_type_abb'] = street_type
counts_cleaned
```

```python
counts_cleaned['street_type'] = [brief_to_full[k] for k in counts_cleaned['street_type_abb']]
counts_cleaned['street_type']
```

```python
counts_cleaned.dropna(subset=['street_type'],inplace=True)
```

```python
counts_cleaned
```

```python
count = np.ones(counts_cleaned.shape[0])
counts_cleaned['count'] = count
counts_cleaned
```

```python
grouped_collision=counts_cleaned.groupby('street_name').sum()
grouped_collision
```

```python
grouped_collision['ratio'] = grouped_collision['count']/grouped_collision['total_count']
grouped_collision
```

```python
grouped_collision.sort_values('ratio', ascending = False)
```
